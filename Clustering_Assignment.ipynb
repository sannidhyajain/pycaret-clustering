## **1. Using K-Mean Clustering**
*   Evaluation Parameters for clusters =3,4,5,6
*   Elbow plot for visualization


  








**1.1 Without Preprocessing**

```python
setup(data = heart_failure , verbose = False)
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  x= create_model('kmeans',num_clusters = i)
  # plot_model(x,plot='tsne')
```

**1.2 Using z-score Normalization**

```python
setup(data = heart_failure ,  verbose = False , normalize=True , normalize_method='zscore')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6,7]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**1.3 Using Transformation**

```python
setup(data = heart_failure ,  verbose = False , transformation=True , transformation_method='yeo-johnson')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**1.4 Using PCA**

```python
setup(data = heart_failure , verbose=False , pca= True , pca_method='linear')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**1.5 Transformation + Normalization**

```python
setup(data = heart_failure , normalize = True , normalize_method='zscore' ,
                             transformation =True , transformation_method= 'yeo-johnson', verbose = False)
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters=i)
```

**1.6 Transformation + Normalization + PCA**

```python
setup(data = heart_failure , transformation= True , transformation_method='yeo-johnson',
                             normalize = True , normalize_method= 'zscore',
                             pca=True , pca_method='linear' , verbose=False)
plot_model(create_model('kmeans'), plot = 'elbow')


c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters=i)
```
